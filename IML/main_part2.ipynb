### Часть 2: для каких задач подходит KMeans
Ниже показаны некоторые примеры использования функции KMeans.

```python
from mylib import *
```

```python
num = 100
x = np.linspace(0, 40, num).reshape(num, 1)
ysin = np.sin(x).reshape(num, 1)
ycos1 = (np.cos(x) + 3).reshape(num, 1)
ycos2 = (np.cos(x) + 2).reshape(num, 1)
matr1 = np.vstack([np.hstack([x, ysin]), np.hstack([x, ycos1])])
matr2 = np.vstack([np.hstack([x, ysin]), np.hstack([x, ycos2])])
cl1, _, _ = KMeans4(matr1, k = 2, strat = "uniform")
cl2, _, _ = KMeans4(matr2, k = 2, strat = "uniform")
pylab.figure(0, figsize = (15, 5))
pylab.subplot(121)
plt.scatter(matr1[:, 0], matr1[:, 1], c = cl1)
pylab.subplot(122)
plt.scatter(matr2[:, 0], matr2[:, 1], c = cl2)
plt.show()
```

```python
matr = np.hstack([x, x])
y = np.zeros(num)
for i in range(16):
    y = x * (i + 2)
    matr = np.vstack([matr, np.hstack([x, y])])
pylab.figure(0, figsize = (8, 8))
cl, _, _ = KMeans4(matr, k = 4, strat = "uniform")
plt.scatter(matr[:, 0], matr[:, 1], c = cl)
plt.show()
```

```python
matr = np.hstack([x, x])
for i in range(4):
    y = x * (i + 2)
    matr = np.vstack([matr, np.hstack([x, y])])
pylab.figure(0, figsize = (8, 8))
cl, _, _ = KMeans4(matr, k = 4, strat = "uniform")
plt.scatter(matr[:, 0], matr[:, 1], c = cl)
plt.show()
```

```python
num = 30
x = np.linspace(20, 40, num).reshape(num, 1)
matr = np.hstack([x, x])
for i in range(3):
    y = x * (i + 2)
    matr = np.vstack([matr, np.hstack([x, y])])
pylab.figure(0, figsize = (8, 8))
cl, _, _ = KMeans4(matr, k = 4, strat = "choice")
plt.scatter(matr[:, 0], matr[:, 1], c = cl)
plt.show()
```

```python
num = 30
x = np.linspace(140, 170, num).reshape(num, 1)
matr = np.hstack([x, x])
for i in range(3):
    y = x * (i + 2)
    matr = np.vstack([matr, np.hstack([x, y])])
pylab.figure(0, figsize = (8, 8))
cl, _, _ = KMeans4(matr, k = 4, strat = "uniform")
plt.scatter(matr[:, 0], matr[:, 1], c = cl)
plt.show()
```

Эта функция лучше всего подходит для нахождения распределения групп точек для групп, которые находятся достаточно далеко друг от друга. Если же искомые группы близки или пересекаются, распределение скорее всего будет неверным.
