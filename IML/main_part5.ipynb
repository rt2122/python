### Часть 5: визуализация матрицы попарных расстояний

```python
from mylib import *
```

```python
n = 2000
k = 4
X, _ = make_blobs(n_samples = n, n_features = 2, centers = k, random_state = 4)
cl, xc, yc = KMeans4(X, k = k, strat = "uniform")
plt.scatter(X[:, 0], X[:, 1], c = cl)
plt.show()
```

```python
matr_dist = make_matr_dist(X)
plt.imshow(matr_dist, cmap='inferno')
plt.show()
```

```python
X, _, _, _ = sort_by_clust(X, cl, xc, yc, corner_left_lower = [True, False])
matr_dist = make_matr_dist(X)
plt.imshow(matr_dist, cmap='inferno')
plt.show()
```

Кластеры отсортированы в зависимости от близости их центров к определенной точке (в данном случае к точке с минимальной абсциссой и максимальной ординатой, то есть к левому верхнему углу прямоугольнника, в который заключены все данные точки).
